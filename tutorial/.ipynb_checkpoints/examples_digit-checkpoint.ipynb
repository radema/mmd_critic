# Example: Digits dataset

```python
import sys
sys.path.append('../')
from mmdcritic.myfunctions import *

from sklearn.datasets import load_digits
import pandas as pd
X = load_digits()
```

```python
feature_names = X['feature_names']
target = X['target']
data = X['data']/16.0
target_names = X['target_names']
```

```python
from sklearn.metrics import pairwise_distances_argmin

prototypes = find_prototypes(data, m = 10)[0]

index_prototypes = pairwise_distances_argmin(prototypes, Y=data, metric='cosine')
```

```python
import matplotlib.pyplot as plt
plt.gray()
for i in index_prototypes:
    plt.matshow(X.images[i])

plt.show()
```
