# Example: Wine Dataset

```python
from sklearn.datasets import load_wine
import pandas as pd
X = load_wine()
```

```python
import sys
sys.path.append('../')
from mmdcritic.myfunctions import *
```

```python
data = X['data']
target = X['target']
target_names = X['target_names']
features_names = X['feature_names']
```

```python
from sklearn.preprocessing import RobustScaler

transformer = RobustScaler(
                          ).fit(data)
```

```python
mmd_result = mmd_critic(data, n_prototypes = 3, n_critics = 3, record = True)
```

```python
prototypes = pd.DataFrame(transformer.transform(mmd_result[0][0]), columns = features_names)
critics = pd.DataFrame(transformer.transform(mmd_result[1][0]), columns = features_names)
```

```python
df = pd.DataFrame(
    transformer.transform(data)
    , columns = features_names)
df['target'] = target
```

```python
import matplotlib.pyplot as plt

wine_to_study = 0

features = [*features_names, features_names[0]]
label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(features))

plt.figure(figsize=(8, 8))
plt.subplot(polar=True)

for i in range(0,len(prototypes)):
    print()
    plt.plot(label_loc, 
             [*prototypes.iloc[i,:].to_list(),prototypes.iloc[i,0]]
             , label='Prototype '+str(i)
            )

plt.plot(
    label_loc,
    [*df.iloc[wine_to_study,:-1].to_list(),df.iloc[wine_to_study,0]]
    , '--'
    , label='Wine to study'
)


plt.title('Prototype comparison', size=20, y=1.05)
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=features)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import pairwise_distances_argmin_min
data2prototypes = pairwise_distances_argmin_min(df.iloc[:,:-1], Y=prototypes, metric='euclidean')
data2critics = pairwise_distances_argmin_min(df.iloc[:,:-1], Y=critics, metric='euclidean')
```

```python
dftmp = pd.DataFrame()

dftmp['mapped_prototype'] = data2prototypes[0].copy()
dftmp['distance_prototype'] = data2prototypes[1].copy()
dftmp['target'] = df['target'].copy()

dftmp['mapped_critic'] = data2critics[0]
dftmp['distance_critic'] = data2critics[1]

n_prototypes = len(prototypes)
di = dftmp.sort_values(by='distance_prototype')[
    ['target','mapped_prototype']
].head(n_prototypes).set_index('mapped_prototype').to_dict()['target']

del dftmp
df['mapped_prototype'] = data2prototypes[0]
df['mapped_target'] = df['mapped_prototype'].map(di)
df['distance_critic'] = data2critics[1]
df['distance_prototype'] = data2prototypes[1].copy()
```

```python
import numpy as np

df.groupby(['target','mapped_target']).agg({'distance_prototype':[np.mean, np.std, 'count'],
                                              'distance_critic': [np.mean, np.std, 'count']})
```

```python
from sklearn.metrics import precision_score

precision_score(df['target'],df['mapped_target'], average = 'weighted')
```

```python
from sklearn.metrics import multilabel_confusion_matrix

multilabel_confusion_matrix(df['target'],df['mapped_target'])
```

```python
dftmp = pd.DataFrame()

dftmp = df[['target','mapped_target','distance_critic','distance_prototype']].copy()
dftmp['correct_mapping'] = (dftmp.target == dftmp.mapped_target).map(lambda x: 'blue' if x else 'orange')
```

```python
dftmp[dftmp.target == 0].plot.scatter(x = 'distance_critic', y = 'distance_prototype', c = 'correct_mapping')
```
