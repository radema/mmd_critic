```python
import sys
sys.path.append('../')
from mmdcritic.myfunctions import *

from sklearn.datasets import load_iris
import pandas as pd
X = load_iris()
```

```python
feature_names = X['feature_names']
target = X['target']
data = X['data']
target_names = X['target_names']
```

```python
df = pd.DataFrame(data, columns = feature_names)

vfunc = np.vectorize(lambda x: target_names[x])

df['target_class'] = vfunc(target)

df = df.sample(frac = 1)
```

```python
import seaborn as sns

g = sns.pairplot(df[feature_names+ ['target_class']], hue = 'target_class', palette = sns.color_palette("Paired",3))
g.map_lower(sns.kdeplot, levels=6, color=".1")
```

```python
from sklearn.metrics import pairwise_distances_argmin_min

mmd_result = mmd_critic(data, n_prototypes = 3, n_critics = 2, record = False)
prototypes = pd.DataFrame(mmd_result[0][0], columns = feature_names)
critics = pd.DataFrame(mmd_result[1][0], columns = feature_names)

data2prototypes = pairwise_distances_argmin_min(df.iloc[:,:-1], Y=prototypes, metric='euclidean')
data2critics = pairwise_distances_argmin_min(df.iloc[:,:-1], Y=critics, metric='euclidean')

df['prototype_flg'] = np.vectorize(lambda x: x<= 1e-06)(data2prototypes[1])
df['critic_flg'] = np.vectorize(lambda x: x<= 1e-06)(data2critics[1])

```

```python
df['role'] = df['prototype_flg'].map(lambda x: 'Prototype' if x==True else '' ) + df['critic_flg'].map(lambda x: 'Critic' if x==True else '' )
df['role'] = df['role'].replace({'':np.NaN}).fillna(df['target_class'])
```

```python
sns.pairplot(df[feature_names + ['role']], 
             hue = 'role', 
             markers=["o", "o", "o",'D','D'],
             palette = sns.color_palette("Paired",5)
            )
```
